```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
# parameters
axx= -0.2; axy= -0.4; bxx= +0.5; bxy= -0.2; cxx= +0.3; cxy= +0.7
dxx= -0.1; dxy= +0.4; fxy=0; gxy=0.5
ayy= -0.2; ayx= +0.6; byy= -0.5; byx= +0.6; cyy= -0.3; cyx= -0.1
dyy= -0.1; dyx= -0.4; fyx=0; gyx=0.7
xi0=-0.0; yi0=-0.0; xp0=+0.0; yp0=+0.0



# equivalent MATLAB code

#clear all; clc; close all;
# % Input the parameters and initial conditions
# axx= -0.2; axy= -0.4; bxx= +0.5; bxy= -0.2; cxx= +0.3; cxy= +0.7;
# dxx= -0.1; dxy= +0.4; fxy=0; gxy=0.5;
# ayy= -0.2; ayx= +0.6; byy= -0.5; byx= +0.6; cyy= -0.3; cyx= -0.1;
# dyy= -0.1; dyx= -0.4; fyx=0; gyx=0.7;
# xi0=-0.0; yi0=-0.0; xp0=+0.0; yp0=+0.0;
```

```python
# create matrix A and vector B

A = np.array([[axx, axy, bxx, bxy], [ayx, ayy, byx, byy], [cxx, cxy, dxx, dxy], [cyx, cyy, dyx, dyy]])
B = np.array([fxy, fyx, gxy, gyx])



# equivalent MATLAB code

# A=[axx axy bxx bxy;ayx ayy byx byy;cxx cxy dxx dxy;cyx cyy dyx dyy]; 
# B=[fxy;fyx;gxy;gyx];
```

```python
#solve the system of ODEs

vdp1 = lambda T, x: A.dot(np.array([x[0], x[1], x[2], x[3]])) + B
sol = solve_ivp (vdp1, [0, 4], np.array([xi0, yi0, xp0, yp0]), max_step=0.1)

[t, xa] = [sol.t, sol.y]



# equivalent MATLAB code

# f = @(t,x) A*[x(1);x(2);x(3);x(4)]+B;
# [t,xa] = ode45(f,[0 4],[xi0 yi0 xp0 yp0]);
```
